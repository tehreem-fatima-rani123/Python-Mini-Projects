```python
import random
import time

# 🎯 Leaderboard dictionary
leaderboard = {}

# 🎨 Encouragement Messages
encouragement = [
    "You're getting closer... or are you? 😈",
    "The shadows are whispering your name... try again!",
    "I see fear in your eyes... dare to guess again?",
    "Tick... Tock... Time is running out! ⏳",
    "What was that sound? Oh, just your wrong guess! 👁️"
]

# 🎮 Start Game Function
def play_game():
    print("\n🎯 Welcome to the Haunted Guessing Game! ☠️\n")

    # 🎚️ Difficulty Selection
    difficulties = {"1": (10, "Easy"), "2": (50, "Medium"), "3": (100, "Hard")}
    for key, value in difficulties.items():
        print(f"{key}. {value[1]} (1-{value[0]})")

    while True:
        difficulty_choice = input("\nChoose difficulty (1/2/3): ")
        if difficulty_choice in difficulties:
            max_number = difficulties[difficulty_choice][0]
            break
        else:
            print("❌ Invalid choice! Try again.")

    # 👁️ Number Selection
    number = random.randint(1, max_number)
    attempts = 0
    start_time = time.time()

    print(f"\nI have chosen a number between 1 and {max_number}. Try to guess! 👁️👁️")

    while True:
        try:
            guess = int(input("\nEnter your guess: "))
            attempts += 1

            if guess < number:
                print(f"📉 Too low! {random.choice(encouragement)}")
            elif guess > number:
                print(f"📈 Too high! {random.choice(encouragement)}")
            else:
                total_time = round(time.time() - start_time, 2)
                print(f"\n🎉 Congratulations! You guessed it in {attempts} attempts and {total_time} seconds. 🏆")

                # 🏆 Save to Leaderboard
                name = input("Enter your name for the leaderboard: ")
                if name:
                    leaderboard[name] = {"attempts": attempts, "time": total_time}

                break
        except ValueError:
            print("❌ Please enter a valid number!")

# 🏆 Display Leaderboard
def show_leaderboard():
    if leaderboard:
        print("\n🏆 LEADERBOARD 🏆")
        sorted_leaderboard = sorted(leaderboard.items(), key=lambda x: (x[1]["attempts"], x[1]["time"]))
        for rank, (player, data) in enumerate(sorted_leaderboard, 1):
            print(f"{rank}. {player} - {data['attempts']} attempts, {data['time']} sec")
    else:
        print("\n🚫 No scores yet!")

# 🎭 Main Loop
while True:
    print("\n🔮 1. Play Game")
    print("📜 2. View Leaderboard")
    print("🚪 3. Exit")

    choice = input("\nEnter your choice: ")

    if choice == "1":
        play_game()
    elif choice == "2":
        show_leaderboard()
    elif choice == "3":
        print("\nGoodbye, mortal! 👁️👁️")
        break
    else:
        print("❌ Invalid choice! Try again.")
```
